```python
#Tugas dan Latihan
# 1 Antrian Bioskop
class AB:
    def __init__(self, kapasitas_maks):
        self.kapasitas_maks = kapasitas_maks
        self.antrian = []

    # Metode untuk menambahkan pelanggan ke dalam antrian
    def enqueue(self, kode_pelanggan):
        if self.size() < self.kapasitas_maks:
            self.antrian.append(kode_pelanggan)
            print(f"{kode_pelanggan} telah ditambahkan ke antrian.")
        else:
            print("Antrian sudah penuh. Harap tunggu sampai ada slot kosong.")

    # Metode untuk melayani pelanggan pertama dalam antrian
    def dequeue(self):
        if not self.is_empty():
            pelanggan = self.antrian.pop(0)
            print(f"{pelanggan} telah dilayani.")
            return pelanggan
        else:
            print("Antrian kosong. Tidak ada pelanggan yang bisa dilayani.")
            return None

    # Metode untuk menampilkan pelanggan pertama dalam antrian
    def front(self):
        if not self.is_empty():
            return self.antrian[0]
        else:
            return "Antrian kosong."

    # Metode untuk menampilkan pelanggan terakhir dalam antrian
    def rear(self):
        if not self.is_empty():
            return self.antrian[-1]
        else:
            return "Antrian kosong."

    # Metode untuk memeriksa apakah antrian kosong
    def is_empty(self):
        return len(self.antrian) == 0

    # Metode untuk menghitung jumlah pelanggan dalam antrian
    def size(self):
        return len(self.antrian)

    # Metode untuk menambahkan pelanggan VIP ke dalam antrian
    def enqueue_vip(self, kode_pelanggan):
        if '7' in kode_pelanggan:
            self.antrian.insert(0, kode_pelanggan)
            print(f"{kode_pelanggan} telah ditambahkan sebagai pelanggan VIP.")
        else:
            self.enqueue(kode_pelanggan)

# Fungsi utama untuk menjalankan program
def main():
    kapasitas_maks = int(input("Masukkan kapasitas maksimal antrian: "))
    bioskop = AB(kapasitas_maks)

    while True:
        print("\nPilihan Menu:")
        print("1. Tambah Pelanggan")
        print("2. Tambah Pelanggan VIP")
        print("3. Layani Pelanggan")
        print("4. Lihat Pelanggan Pertama")
        print("5. Lihat Pelanggan Terakhir")
        print("6. Lihat Jumlah Pelanggan")
        print("7. Keluar")

        pilihan = input("Masukkan pilihan (1-7): ")

        if pilihan == "1":
            kode_pelanggan = input("Masukkan kode pelanggan: ")
            bioskop.enqueue(kode_pelanggan)
        elif pilihan == "2":
            kode_pelanggan = input("Masukkan kode pelanggan: ")
            bioskop.enqueue_vip(kode_pelanggan)
        elif pilihan == "3":
            bioskop.dequeue()
        elif pilihan == "4":
            print(f"Pelanggan pertama: {bioskop.front()}")
        elif pilihan == "5":
            print(f"Pelanggan terakhir: {bioskop.rear()}")
        elif pilihan == "6":
            print(f"Jumlah pelanggan: {bioskop.size()}")
        elif pilihan == "7":
            print("Terima kasih telah menggunakan sistem antrian bioskop.")
            break
        else:
            print("Pilihan tidak valid. Harap masukkan angka 1-7.")

if __name__ == "__main__":
    main()
```

```python
# 2 Antrean Fotokopi Kampus
class PhotocopyQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Pesanan {order_id} dengan {num_pages} halaman diproses segera (Fast Track).")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} dengan {num_pages} halaman ditambahkan ke antrian.")

    def dequeue(self):
        if not self.is_empty():
            order_id, num_pages = self.queue.pop(0)
            if num_pages > 100:
                remaining_pages = num_pages - 100
                self.enqueue(order_id, remaining_pages)
                print(f"Pesanan {order_id} diproses untuk 100 halaman. Sisa {remaining_pages} halaman ditambahkan kembali ke antrian.")
            else:
                print(f"Pesanan {order_id} diproses untuk {num_pages} halaman.")
        else:
            print("Antrian kosong. Tidak ada pesanan untuk diproses.")

    def front(self):
        if not self.is_empty():
            return self.queue[0]
        else:
            return "Antrian kosong."

    def rear(self):
        if not self.is_empty():
            return self.queue[-1]
        else:
            return "Antrian kosong."

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = PhotocopyQueue()
queue.enqueue("FAI22", 50)
queue.enqueue("RAM18", 120)
queue.enqueue("AKM47", 8)
queue.enqueue("FER07", 200)

print("Pesanan di depan:", queue.front())
print("Pesanan di belakang:", queue.rear())

queue.dequeue()
queue.dequeue()

print("Ukuran antrian:", queue.size())

queue.dequeue()
queue.dequeue()
```

```python
# 3 Antrean Parkir Mall
# Mendefinisikan kelas Queue untuk antrian parkir
class AP:
    def __init__(self):
        self.items = []  # Inisialisasi antrian dengan daftar kosong

    def enqueue(self, item):
        # Menambahkan kendaraan ke dalam antrian (masuk parkir)
        self.items.append(item)

    def dequeue(self):
        # Mengeluarkan kendaraan yang lebih dulu datang (keluar parkir)
        if not self.is_empty():
            return self.items.pop(0)
        return None

    def front(self):
        # Menampilkan kendaraan pertama dalam antrian
        if not self.is_empty():
            return self.items[0]
        return None

    def rear(self):
        # Menampilkan kendaraan terakhir dalam antrian
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        # Mengecek apakah antrian kosong
        return len(self.items) == 0

    def size(self):
        # Menghitung jumlah kendaraan dalam antrian
        return len(self.items)


# Inisialisasi antrian parkir
parkir = AP()

# Kapasitas maksimum parkir
kapasitas_parkir = 15

# Contoh nomor plat kendaraan (gunakan dua huruf pertama kota lahir + dua digit terakhir NPM)
kendaraan_biasa = ["KE22", "BD34", "SB56"]
kendaraan_listrik = ["EL78", "EK90"]

# Fungsi untuk menambahkan kendaraan ramah lingkungan dengan prioritas
def enqueue_priority(vehicle):
    # Menambahkan kendaraan ramah lingkungan ke depan antrian
    parkir.items.insert(0, vehicle)  # Menempatkan kendaraan listrik di depan antrian


# Menambahkan kendaraan ke dalam antrian parkir
for kendaraan in kendaraan_biasa:
    if parkir.size() < kapasitas_parkir:
        parkir.enqueue(kendaraan)
    else:
        print(f"Parkir penuh, kendaraan {kendaraan} menunggu...")

# Menambahkan kendaraan ramah lingkungan dengan prioritas
for kendaraan in kendaraan_listrik:
    if parkir.size() < kapasitas_parkir:
        enqueue_priority(kendaraan)
    else:
        print(f"Parkir penuh, kendaraan ramah lingkungan {kendaraan} menunggu...")

# Menampilkan kendaraan pertama dan terakhir dalam antrian
print(f"Kendaraan pertama dalam antrian: {parkir.front()}")
print(f"Kendaraan terakhir dalam antrian: {parkir.rear()}")

# Mengeluarkan beberapa kendaraan dari antrian
print(f"Kendaraan keluar: {parkir.dequeue()}")
print(f"Kendaraan keluar: {parkir.dequeue()}")

# Menampilkan jumlah kendaraan dalam antrian
print(f"Jumlah kendaraan dalam antrian: {parkir.size()}")

# Mengecek apakah antrian parkir kosong
print(f"Apakah antrian parkir kosong? {'Ya' if parkir.is_empty() else 'Tidak'}")
```

```python
# 4 Antean Pengambilan Sertifikat kampus
class SK:
    def __init__(self):
        self.queue = []

    def enqueue(self, student_id):
        self.queue.append(student_id)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None

    def front(self):
        if not self.is_empty():
            return self.queue[0]
        return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1]
        return None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def split_queue(self):
        regular_queue = []
        achievement_queue = []
        for student in self.queue:
            if self.is_achievement_student(student):
                achievement_queue.append(student)
            else:
                regular_queue.append(student)
        return regular_queue, achievement_queue

    def is_achievement_student(self, student_id):
        # Implementasi logika untuk mengecek apakah mahasiswa pernah mengikuti kompetisi tingkat nasional
        # Misalnya, kita bisa menggunakan daftar mahasiswa berprestasi
        achievement_students = ["T22", "I04", "E20"]  # Contoh daftar mahasiswa berprestasi
        return student_id in achievement_students

# Contoh penggunaan
queue = SK()
queue.enqueue("T22")
queue.enqueue("M30")
queue.enqueue("E20")
queue.enqueue("I04")

print("Antrian awal:", queue.queue)
print("Mahasiswa pertama:", queue.front())
print("Mahasiswa terakhir:", queue.rear())
print("Ukuran antrian:", queue.size())

if queue.size() > 20:
    regular_queue, achievement_queue = queue.split_queue()
    print("Jalur mahasiswa reguler:", regular_queue)
    print("Jalur mahasiswa berprestasi:", achievement_queue)
else:
    print("Antrian tidak perlu dibagi.")

# Melayani mahasiswa
print("Melayani mahasiswa:", queue.dequeue())
print("Antrian setelah melayani:", queue.queue)
```
