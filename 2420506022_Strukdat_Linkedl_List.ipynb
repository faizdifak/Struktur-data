```python
# Definisi Node dalam Linked List
class TaskNode:
    def __init__(self, task_name):
        self.task_name = task_name #Menyimpan nama tugas
        self.next = None #Pointer ke tugas berikutnya

#Definisi Linked List untuk Daftar Tugas
class TaskList:
    def __init__(self):
        self.head = None #Awalnya tidak ada tugas (daftar kosong)

   #Menambahkan tugas ke Linked list
    def add_task(self, task_name):
       new_task = TaskNode(task_name)
       if not self.head:
           self.head = new_task #Jika list kosong, tugas pertama menjadi head

       else:
          current = self.head
          while current.next: #Mencari node terakhir dalam list
              current = current.next
          current.next = new_task #Menambahkan tugas di akhir list

    #Menampilkan daftar tugas
    def show_tasks(self):
        if not self.head:
            print("\n Tidak ada tugas hari ini!")
            return

        print("\n Daftar tugas hari ini:")
        current = self.head
        count = 1
        while current:
           print(f"{count}. {current.task_name}")
           current = current.next
           count += 1

# ==== Program Utama(main) ====#

#membuat objek daftar tugas
task_list = TaskList()

#Meminta pengguna masukkan jumlah tugas
num_tasks = int(input("Masukkan jumlah tugas hari ini:"))

#Loop untuk memasukkan tugas ke dalam linked list
for i in range(1, num_tasks + 1):
  task_name = input(f"Masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

#menampilkan hasil daftar tugas
task_list.show_tasks()
```

```python
#Definisi Node dalam Doubly Linked List
class PageNode:
    def __init__(self, url):
        self.url = url #URL halaman web
        self.next = None #Pointer ke halaman berikutnya
        self.prev = None #Pointer ke halaman sebelumnya

#Definisi Doubly Linked list untuk Riwayat Peramban
class BrowserHistory:
    def __init__(self):
        self.current = None #Pointer ke halaman yang sedang dibuka

    #Menambahkan halaman baru ke dalam riwayat
    def visit_page(self, url):
        new_page = PageNode(url)
        if not self.current:
            self.current = new_page #Jika riwayat kosong, halaman ini menjadi halaman pertama
        else:
            new_page.prev = self.current
            self.current.next = new_page
            self.current = new_page #Pindahkan posisi ke halaman baru

    #Menampilkan riwayat kunjungan
    def show_history(self):
        if not self.current:
            print("\n Riwayat kosong.")
            return

        print("\n Riwayat Pramban Web:")
        temp = self.current
        while temp.prev: #Kembali ke halaman pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.url}")
            temp = temp.next
            count += 1

    #Navigasi mundur ke halaman sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\n Anda kembali ke: {self.current.url}")
        else:
            print("\n Tidak bisa kembali, ke halaman pertama!")

    #Navigasi maju ke halaman berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\n Anda maju ke: {self.current.url}")
        else:
            print("\n Tidak bisa maju, ini adalah halaman terakhir!")

# ==== Program Utama (main) ====#

# Membuat objek riwayat perambanan
history = BrowserHistory() # Call the constructor to create an instance

#Meminta pengguna masukkan jumlah halaman yang dikunjungi
num_pages = int(input("\nMasukkan jumlah halaman yang dikunjungi:"))

#Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}:")
  history.visit_page(url)

#Menampilkan riwayat yang telah dimasukkan
history.show_history()

#Menu navigasi pengguna
while True:
    command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
      history.go_back()
    elif command == 'f':
      history.go_forward()
    elif command == 'x':
      print('\nKeluar dari wilayah perambanan.')
      break
    else:
      print("\nPerintah tidak dikenal.")






```

```python
#Latihan2
class FolderNode:
    def __init__(self, folder_name):
        self.folder_name = folder_name  # Menyimpan nama folder
        self.next = None  # Menyimpan referensi ke node berikutnya
        self.prev = None  # Menyimpan referensi ke node sebelumnya

class FileExplorerHistory:
    def __init__(self):
        self.current = None  # Menyimpan referensi ke folder saat ini

    def visit_folder(self, folder_name):
        new_folder = FolderNode(folder_name)  # Membuat node folder baru
        if not self.current:  # Jika belum ada folder saat ini
            self.current = new_folder  # Jadikan folder baru sebagai folder saat ini
        else:
            new_folder.prev = self.current  # Hubungkan folder baru dengan folder saat ini
            self.current.next = new_folder  # Hubungkan folder saat ini dengan folder baru
            self.current = new_folder  # Perbarui folder saat ini menjadi folder baru

    def show_history(self):
        if not self.current:
            print("\nRiwayat navigasi kosong.")  # Tampilkan pesan jika riwayat kosong
            return

        print("\nRiwayat Navigasi File Explorer:")
        temp = self.current  # Mulai dari folder saat ini
        while temp.prev:  # Kembali ke folder pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.folder_name}")  # Tampilkan nama folder dan urutannya
            temp = temp.next
            count += 1

    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev  # Pindah ke folder sebelumnya
            print(f"\nAnda kembali ke: {self.current.folder_name}")
        else:
            print("\nTidak bisa kembali, ini adalah folder pertama!")  # Pesan jika tidak bisa kembali

    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next  # Pindah ke folder berikutnya
            print(f"\nAnda maju ke: {self.current.folder_name}")
        else:
            print("\nTidak bisa maju, ini adalah folder terakhir!")  # Pesan jika tidak bisa maju

#=== Program Utama (Main) ====#

history = FileExplorerHistory()

num_folders = int(input("Masukkan jumlah folder yang ingin dinavigasikan: "))  # Minta jumlah folder

for i in range(1, num_folders + 1):
    folder_name = input(f"Masukkan nama folder ke-{i}: ")  # Minta nama folder
    history.visit_folder(folder_name)  # Kunjungi folder tersebut

history.show_history()  # Tampilkan riwayat navigasi

while True:
    command = input("\nTekan 'b' untuk kembali, 'f' untuk maju, 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back()  # Panggil fungsi untuk kembali ke folder sebelumnya
    elif command == 'f':
        history.go_forward()  # Panggil fungsi untuk maju ke folder berikutnya
    elif command == 'x':
        print('\nKeluar dari navigasi File Explorer.')
        break  # Keluar dari loop dan akhiri program
    else:
        print("\nPerintah tidak dikenal.")  # Pesan jika perintah tidak dikenal
```

```python
#Definisi Node dalam Circular Linked list
class SongNode:
    def __init__(self, title):
      self.title = title #Menyimpan judul lagu
      self.next = None #Pointer ke lagu berikutnya

#Definisi Circular Linked list untuk playlist judul lagu
class Playlist:
    def __init__(self):
      self.current = None #Lagu yang sedang diputar

    #Menambahkan lagu ke dalam playlist  #Corrected indentation here
    def add_song(self, title):
        new_song = SongNode(title)
        if not self.current:
           self.current = new_song
           self.current.next = self.current #Menghubungkan ke dirinya sendiri(circular)

        else:
            temp = self.current
            while temp.next != self.current: #Mencari lagu terakhir
             temp = temp.next
            temp.next = new_song #Menambahkan lagu diakhir
            new_song.next = self.current #Menghubungkan kembali ke awal

      #Menampilkan daftar lagu dalam playlist
    def show_playlist(self):
        if not self.current:
          print("\nPlaylist kosong!")
          return

        print("\nPlaylist lagu anda:")
        temp = self.current
        count = 1
        while True:
            print(f"{count}. {temp.title}")
            temp = temp.next
            count += 1
            if temp == self.current: #Berhenti jika sudah kembali ke awal
               break

      #Memutar lagu berikutnya dalam playlist
    def play_next(self): #Renamed next_song to play_next
          if self.current:
            self.current = self.current.next
            print(f"\nMemutar:{self.current.title}")

# ==== Program Utama (main) ====#

#Membuat objek playlist
playlist = Playlist()

#Meminta pengguna memasukkan jumlah lagu
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: "))

#Loop untuk memasukan lagu ke dalam circular linked list
for i in range(1, num_songs + 1):
    title = input(f"Masukkan judul lagu ke-{i}:")
    playlist.add_song(title)

#Menampilkan daftar lagu
playlist.show_playlist()

#Menu navigaasi pengguna
while True:
  command = input("\nTekan 'n' untuk lagu berikutnya, tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
  if command == 'n':
    playlist.play_next() # Calling the correct method name
  elif command == 'x':
    print("\nKeluar dari pemutar musik.")
    break
  else:
    print("\nPerintah tidak dikenal.")

```

```python
#Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data #Menyimpan nilai node
        self.next = None #Pointer ke node berikutnya

#Definisi Singly Lingked List dengan berbagai operasi
class LinkedList:
    def __init__(self):
        self.head = None #Awalnya, daftar kosong

    #Menambah elemen diawal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    #Menambah elemen diakhir
    def add_at_end(self, data):
        new_node =Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    #Menambah elemen di posisi tertentu
    def add_atposition(self, position, data):
        if position < 1:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
          self.add_at_beginning(data)
          return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    #Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    #Menghapus elemen akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    #Menghapus elemen diposisi tertentu
    def delate_ata_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position -1):
            if not temp or not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    #Mencari elemen dalam linked list
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
               print(f"Elemen{key}ditemukan dalam daftar!")
               return
            temp = temp.next
        print(f"Elemen{key}tidak ditemukkan dalam daftar.")

    #Membalik (reverse)Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self. head = prev

    #Menampilkan Link List
    def display(self):
        if not self.head:
            print("Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# ==== Program Utama (main) ====#

#Membuat objek link list
ll = LinkedList()

#Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"\nMasukkan data ke-{i+1}: "))
    ll.add_at_end(data)

#Menu Operasi Linked List
while True:
    print("\nPilih Operasi:")
    print("1. Tambah Elemen Diawal")
    print("2. Tambah Elemen Diakhir")
    print("3. Tambah Elemen Di Posisi Tertentu")
    print("4. Hapus Elemen Diawal")
    print("5. Hapus Elemen Di Akhir")
    print("6. Hapus Elemen Di Posisi Tertentu")
    print("7. Cari Elemen dalam Daftar")
    print("8. Balikkan Daftar")
    print("9. Tampilkan Daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan:"))

    if choice == 1:
        data = int(input("\nMasukkan data yang ingin ditambahkan diawal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
       data = int(input("\nMasukkan data yang ingin ditambahkan diakhir:"))
       ll.add_at_end(data)
    elif choice == 3:
       position = int(input("\nMasukkan posisi untuk menambahkan data: "))
       data = int(input("\nMasukkan data yang ingin ditambahkan:"))
       ll.add_atposition(position, data)
    elif choice == 4:
       ll.delete_at_beginning()
    elif choice == 5:
       ll.delete_at_end()
    elif choice == 6:
       position = int(input("\nMasukkan posisi elemen yang ingin dihapus:"))
       ll.add_at_position(position, data)
    elif choice == 7:
       key = int(input("\nMasukkan elemen yang ingin dicari:"))
       ll.search(key)
    elif choice == 8:
       ll.reverse()
       print("Daftar berhasil dibalik!")
    elif choice == 9:
       ll.display()
    elif choice == 0:
       print("\nKeluar dari program.")
       break
    else:
       print("\nPilihan tidak valid, silakan coba lagi!")







```

```python
#Latihan 3
#Operasi dasar dari program diatas
#Menambahkan elemen di awal
def add_at_beginning(self, data):
    new_node = Node(data)  # Buat node baru
    new_node.next = self.head  # Hubungkan node baru ke head saat ini
    self.head = new_node  # Jadikan node baru sebagai head
#Menambahkan elemen di akhir
def add_at_end(self, data):
    new_node = Node(data)  # Buat node baru
    if not self.head:  # Jika daftar kosong, jadikan node baru sebagai head
        self.head = new_node
        return
    temp = self.head
    while temp.next:  # Telusuri hingga node terakhir
        temp = temp.next
    temp.next = new_node  # Tambahkan node baru di akhir
#Menambahkan elemen diposisi tertentu
def add_at_position(self, position, data):
    if position < 0:
        print("Posisi tidak valid!")
        return
    new_node = Node(data)
    if position == 0:  # Jika posisi awal, panggil add_at_beginning
        self.add_at_beginning(data)
        return
    temp = self.head
    for _ in range(position - 1):  # Telusuri sampai sebelum posisi yang diinginkan
        if not temp:
            print("Posisi melebihi panjang list!")
            return
        temp = temp.next
    if not temp:
        print("Posisi melebihi panjang list!")
        return
    new_node.next = temp.next
    temp.next = new_node
#Menghapus elemen di awal
def delete_at_beginning(self):
    if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus!")
        return
    self.head = self.head.next  # Pindahkan head ke node berikutnya
#Menghapus elemen di akhir
def delete_at_end(self):
    if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus!")
        return
    if not self.head.next:  # Jika hanya ada satu node
        self.head = None
        return
    temp = self.head
    while temp.next.next:  # Telusuri sampai node kedua terakhir
        temp = temp.next
    temp.next = None  # Hapus node terakhir
#Menghapus diposisi tertentu
def delete_at_position(self, position):
    if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus!")
        return
    if position == 0:
        self.head = self.head.next  # Hapus elemen pertama
        return
    temp = self.head
    for _ in range(position - 1):  # Telusuri hingga node sebelum yang akan dihapus
        if not temp or not temp.next:
            print("Posisi tidak ditemukan!")
            return
        temp = temp.next
    temp.next = temp.next.next  # Lewati node yang akan dihapus
#Mencari elemen dalam Linked List
def search(self, key):
    temp = self.head
    while temp:  # Telusuri semua elemen
        if temp.data == key:
            print(f"Elemen {key} ditemukan dalam daftar!")
            return
        temp = temp.next
    print(f"Elemen {key} tidak ditemukan dalam daftar.")
#Membalik Linked List
def reverse(self):
    prev = None
    current = self.head
    while current:
        next_node = current.next  # Simpan node berikutnya
        current.next = prev  # Balik arah pointer
        prev = current
        current = next_node
    self.head = prev  # Atur head ke node terakhir
```
